```python
import requests
```

```python
url = 'https://jobs.github.com/positions.json'
technologies = ['C', 'C#','C++','Java', 'JavaScript'
,'Python'
,'Scala'
,'Oracle'
,'SQL Server'
,'MySQL Server'
,'PostgreSQL'
,'MongoDB']
for _ in technologies:
    pageNumber = 0
    payload={"description":_, 'location':'', 'page': 0}
    r = requests.get(url, params = payload)
    totalJobs = 0
    while len(r.json()) != 0:
        payload={"description":_, 'location':'', 'page': pageNumber}
        r = requests.get(url, params = payload)
        Jobs = len(r.json())
        totalJobs += Jobs
        pageNumber+=1
    print(f'Technology:{_} Jobs Available:{totalJobs}')
```
